# Visuals 

## - Create a bar chart showing the top 10 happiest countries and their happiness scores. 

Not sure if this means of each year or maybe score of all the years, I do think I will be way better to have the score considering all years like a mean. 
So A will be each year and B a mean of Score over the years we have 

### A

```python
df_all['Rank'] = df_all.groupby('Year')['Score'].rank(ascending=False)

top10_by_year = df_all[df_all['Rank'] <= 10]

g = sns.FacetGrid(top10_by_year, col="Year", col_wrap=2, height=5, sharex=False)
g.map_dataframe(sns.barplot, x="Score", y="Country", palette='viridis')
g.set_titles("Top 10 in {col_name}")
g.set_axis_labels("Score", "Country")
plt.tight_layout()
plt.show()
```

### B

```python
avg_score_per_country = df_all.groupby('Country')['Score'].mean().reset_index()

top10_countries = avg_score_per_country.nlargest(10, 'Score')

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_countries, x='Score', y='Country', palette='viridis')
plt.title('Top 10 Countries by Average Happiness Score')
plt.xlabel('Average Happiness Score')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

## - (Optional) Create a bar chart for the bottom 10 (least happy) countries. Will be using my B approach

```python
avg_score_per_country = df_all.groupby('Country')['Score'].mean().reset_index()

top10_countries = avg_score_per_country.nsmallest(10, 'Score')

plt.figure(figsize=(10, 6))
sns.barplot(data=top10_countries, x='Score', y='Country', palette='viridis')
plt.title('Top 10 Countries by Average Happiness Score (lowest 10)')
plt.xlabel('Average Happiness Score')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```
